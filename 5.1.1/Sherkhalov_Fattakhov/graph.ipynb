```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lamda = np.array([5401, 5852, 5945, 6143, 6402, 6507])
result = np.array([1892, 2150, 2196, 2288, 2396, 2500])

x = result
y = lamda
dx = 1
dy = 0

plt.xlabel(r'Показания прибора', fontsize=14)
plt.ylabel(r'$\lambda$,\, \AA', fontsize=14)
plt.title(r'График калибровки', fontsize=14)
plt.grid(True)
plt.errorbar(x, y, xerr=dx, yerr=dy, fmt='o', color='black', capsize=3)

pol = np.polyfit(x, y, deg=1, cov=True)
print(pol)

k = pol[0][0]
b = pol[0][1]
dk = pol[1][0][0]
db = pol[1][1][1]

pol = pol[0]
yfit = np.polyval(pol, x)
plt.plot(x, yfit, color="firebrick", label=r'k = '+str(k)+'\n'+'b = '+str(b)+' $\pm$ '+str(db))

plt.legend(loc='best', fontsize=12)
plt.show()
```
