```python
filename = 'afc1.csv'
df = pd.read_csv(filename, delimiter=';')
```

```python
xdata = df.f/32.4
ydata = df.U/1.383

x1 = df.f
y1 = df.U
```

```python
plt.rcParams['figure.dpi'] = 300

style_scatter = '.r'
style_line = '--k'
```

```python
fig = plt.figure(figsize=(6.4, 4.8))
ax = fig.add_subplot(111)
ax.set_xlim((0.90, 1.1))
ax.set_ylim((0.0, 1.1))
ax.set_xlabel(r'$f/f_{01}, 1$')
ax.set_ylabel(r'$U/U_{01}, 1$')

ax.set_xticks(np.arange(0.9, 1.1, 0.02))
ax.set_xticks(np.arange(0.9, 1.1, 0.004), minor=True)

ax.plot((0.9820, 0.9820),(0, 0.707), '-.k', lw=1)
ax.plot((1.0203, 1.0203),(0, 0.707), '-.k', lw=1, label=r'$\frac{\Delta f}{f_0} = 38.3\cdot 10^{-3}$')

ax.plot(xdata, ydata, style_scatter, ms=5, lw=1, label='АЧХ $RLC_1$')
ax.plot((0.9, 1.1), (0.707, 0.707), '--k', label=r'$U/U_0 = \frac {1} {\sqrt{2}}$')

ax.plot(xdata[12:14], ydata[12:14], '-r')
ax.plot(xdata[19:21], ydata[19:21], '-r')

ax.grid(which='major', alpha=0.5)
ax.grid(which='minor', alpha=0.2)

ax.legend()
```
