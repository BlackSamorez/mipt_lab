```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sin_plus = np.array([
0.2028,
0.2183,
0.2463,
0.2736,
0.2891,
0.2901,
0.3068,
0.3125])
sin_minus=np.array([
0.2018,
0.2175,
0.2453,
0.2724,
0.2877,
0.2888,
0.3055,
0.3109])
lam = np.array([
404.7, 
435.8, 
491.6, 
546.1, 
577.0, 
579.1, 
623.4, 
690.7])
```

```python
plot_1 = plt.figure(figsize=(12*0.9,10*0.9))
plt.grid(visible=True, linewidth=0.6)

plt.title(' $sin ()$', fontsize=18)
#plt.xlim(xmin=0, xmax=6)
#plt.ylim(ymin=0, ymax=19)
plt.xlabel('$$,    ', fontsize=20, rotation=0,
           #ha='right'
          )
plt.ylabel('$sin $', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=16, size=10)
plt.grid(visible=True, linewidth=0.6)


yerr_1 = 0.002*sin_plus
#xerr_1 = 0.01

plt.errorbar(
    lam,
    sin_plus,
    fmt='ro',
    linewidth=1,
    markersize=2,
    elinewidth=1,
    label = 'm = +1',
    #xerr=xerr_1,
    yerr=yerr_1,
    capsize=0
)
plt.errorbar(
    lam,
    sin_minus,
    fmt='bo',
    linewidth=1,
    markersize=2,
    elinewidth=1,
    label = 'm = -1',
    #xerr=xerr_1,
    yerr=yerr_1,
    capsize=0
)

z = np.polyfit(lam[:-2], sin_minus[:-2], deg=1)
polynom_1 = np.poly1d(z)
plt.plot(lam, polynom_1(lam),"k--", linewidth=1, label="")


plt.legend(fontsize=18, markerscale = 2)
plt.show()

plot_1.savefig(
    'sin.pdf',
    format='pdf',
    bbox_inches='tight',
    #pad_inches=4
)
```
