```python
def clean_data(df):
    df.tanphi = df.tanphi.apply(lambda x: "{:.3f}".format(x)).astype(float)
    return df


# Loaded variable 'df' from URI: c:\Users\stg05\OneDrive\2_COURSE_2024_2025\labs\3.7.1\2.csv
df = pd.read_csv(
    r"c:\Users\stg05\OneDrive\2_COURSE_2024_2025\labs\3.7.1\2.csv", sep=";"
)

df = clean_data(df.copy())

print(df.to_latex())

xdata = df.hz
ydata = df.tanphi

popt, pcov = curve_fit(curve, xdata, ydata)
k, b = popt
icurve = lambda x: curve(np.array(x), k, b)
```

```python
fig = plt.figure(figsize=(6.4, 4.0))
ax = fig.add_subplot(111)

ax.set_ylim((0, 6))
ax.set_xlim((0, 600))
ax.set_ylabel(r'$\tan (\psi)$')
ax.set_xlabel(r'$f, \mathrm{Hz}$')


ax.set_yticks(np.arange(0, 6, 1))
ax.set_yticks(np.arange(0, 6, .2), minor=True)
ax.set_xticks(np.arange(0, 600, 20), minor=True)
ax.set_xticks(np.arange(0, 600, 100))

ax.plot((0, 600), icurve((0,600)), '--k', lw=1, label=r'$ k = $' + '{:.2f}'.format(k*1000)+r'$\cdot 10^{-3}~\mathrm{Hz}^{-1}$')

ax.plot(xdata, ydata, '.r', ms=5, label='Data')
ax.legend()
ax.grid(which='minor', alpha=0.2)
ax.grid(which='major', alpha=0.5)
k,b
```
