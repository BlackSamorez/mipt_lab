```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

r = np.array((np.mean((31,31,30, 33)), 
np.mean((45,44,43,43)), 
np.mean((53,53,54,55)), 
np.mean((62,61,62,63)), 71, 77.5))

sy = np.array((np.std((31,31,30, 33)),
np.std((45,44,43,43)),      
np.std((53,53,54,55)),
np.std((62,61,62,63)),
np.std((71, 72)),
np.std((78,77))))

sy = np.sqrt(sy**2 + np.ones(6))
sr2 = sy / r * 2 * r**2
no = np.array(range(1, 7))
plt.ylabel(r'$r^2,~\mathrm{mm^2}$')
plt.xlabel(r'$m$')
plt.errorbar(x=no, y=r**2, yerr=sr2,capsize=3, fmt='ro', ms=4,label='data')
plt.grid(True)

curve = lambda x, a, b: a * x + b
popt, pcov = curve_fit(curve, no, r**2)
x = np.linspace(0, 7, 100)
plt.plot(x, curve(x, *popt), 'k--',  linewidth=1,label='fit: k = %.0f$\\pm$%.0f mm$^2$' % (popt[0], np.sqrt(pcov[0][0])))
plt.legend()
```
