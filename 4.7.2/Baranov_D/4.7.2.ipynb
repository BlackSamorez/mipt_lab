```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
m = np.linspace(1,7, 7)
rm = np.array([
2.5,
3.7,
4.5,
5.3,
6.2,
6.8,
7.3])

eps = np.array([
0.040,
0.027,
0.022,
0.019,
0.016,
0.015,
0.014])
```

```python
plot_1 = plt.figure(figsize=(12*0.9,10*0.9))
plt.grid(visible=True, linewidth=0.6)

plt.title('Зависимость $r_m^2(m)$', fontsize=18)
#plt.xlim(xmin=0, xmax=6)
#plt.ylim(ymin=0, ymax=19)
plt.xlabel('$m$, номер тёмного кольца', fontsize=20, rotation=0,
           #ha='right'
          )
plt.ylabel('$r^2$, см$^2$', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=16, size=10)
plt.grid(visible=True, linewidth=0.6)


yerr_1 = rm**2 *2*eps
#xerr_1 = 0.01

#z, err = np.polyfit(D, Q[1], deg=1, cov=True)
#polynom_2 = np.poly1d(z)
#plt.plot(D, polynom_2(D),"b--", linewidth=1, label="МНК")

plt.errorbar(
    m,
    rm**2,
    fmt='ro',
    linewidth=1,
    markersize=4,
    elinewidth=1,
    label = 'Квадрат радиуса тёмного кольца',
    #label='В "одну" сторону',
    #xerr=xerr_1,
    yerr=yerr_1,
    capsize=1
)

z = np.polyfit(m, rm**2, deg=1)
polynom_1 = np.poly1d(z)
plt.plot(np.linspace(0,7,8), polynom_1(np.linspace(0,7,8)),"k--", linewidth=1, label="МНК")


plt.legend(fontsize=18, markerscale = 2)
plt.show()

plot_1.savefig(
    'rm(m).pdf',
    format='pdf',
    bbox_inches='tight',
    #pad_inches=4
)
```
