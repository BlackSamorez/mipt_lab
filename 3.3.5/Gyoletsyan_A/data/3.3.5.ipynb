```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
dB = {1: 5, 2: 1, 3: 1}
dI = 0.5
dU = 0.0002
```

```python
cols = pd.MultiIndex.from_product([["Pair1", "Pair2", "Pair3"], ["B", "dB", "I", "U"]])
idx = pd.MultiIndex.from_product([[5, 6, 7, 8, 9, 10], [0, 1, 2, 3, 4]])
data = pd.DataFrame(columns=cols, index=idx, dtype=float)
```

```python
for i in range(1, 4):
    data_raw = pd.read_csv(f"data{i}.csv")
    pair = f"Pair{i}"
    
    for B in range(5, 11):
        slc = slice((B-5)*5, (B-4)*5 - 1)
        
        data.loc[B, (pair, "B")].loc[:] = data_raw.loc[slc,"B"].to_numpy()
        data.loc[B, (pair, "dB")].loc[:] = dB[i]
        data.loc[B, (pair, "I")].loc[:] = data_raw.loc[slc,"I"].to_numpy()
        data.loc[B, (pair, "U")].loc[:] = (data_raw.loc[slc,"U2"] - data_raw.loc[slc,"U1"]).to_numpy() / 2

data
```

```python
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(projection='3d', aspect="auto")

colors = ["", "red", "blue", "green"]

for i in range(1,4):
    d = data.loc[:, f"Pair{i}"]
    I, B, U = d["I"].to_numpy(), d["B"].to_numpy(), d["U"].to_numpy()
    dB = d["dB"].to_numpy()
    
    I2 = I.reshape((6,5))
    B2 = B.reshape((6,5))
    U2 = U.reshape((6,5))

    ax.scatter(I, B, U, marker='o', s=20, color=colors[i])
    ax.errorbar(I, B, U, xerr=dI, yerr=dB, zerr=dU, linestyle='', color=colors[i])
    ax.plot_wireframe(I2, B2, U2, color=colors[i])
    
    ax.set_title(" $I, U, B$", fontsize=15)
    
    ax.set_xlabel(r'$I, mA$', fontsize=15)
    ax.set_ylabel(r'$B, mT$', fontsize=15)
    ax.set_zlabel(r'$U, mV$', fontsize=15)

plt.savefig("../pictures/plot_IUB.eps")
```

```python
def mnk(X, Y):
    a, b = np.polyfit(X, Y, deg=1)
    
    S_x = np.std(X) ** 2
    S_y = np.std(Y) ** 2
    da = 2 * np.sqrt((S_y/S_x - a ** 2)/(len(X) - 2))
    db = da * np.sqrt(np.mean(X**2))
    
    def f(x):
        return a*x + b
    
    return a, b, da, db, f
```

```python
U = data.loc[:, (slice(None), ["U"])].to_numpy().flatten()/1000 # V
I = data.loc[:, (slice(None), ["I"])].to_numpy().flatten()/1000 # A
B = data.loc[:, (slice(None), ["B"])].to_numpy().flatten()/1000 # T
dB = data.loc[:, (slice(None), ["dB"])].to_numpy().flatten()/1000 # T

h = 50 * 10**-9
```

```python
plt.figure(figsize=(18, 12), dpi=100)
plt.rcParams['font.size'] = 20
plt.grid(True)
plt.title(r"$Uh(IB)$")
plt.xlabel(r"$IB,  \cdot $")
plt.ylabel(r"$Uh,  \cdot $")

X = I*B # A*T
Y = U*h # U*h

dX = X * np.sqrt(((dI/1000)/I)**2 + (dB/B)**2)
dY = (dU/1000)*h

a, b, da, db, f = mnk(X, Y)

plt.scatter(X, Y)
plt.errorbar(X, Y, xerr=dX, yerr=dY, linestyle='')
x_mm = np.array([0, X.max() + 0.005])
plt.plot(x_mm, f(x_mm), color='red', label="$y = (2.813 \pm 0.062)\cdot 10^{-11} x + (5.4 \pm 3.2)\cdot 10^{-14}$")

plt.legend()

plt.savefig("../pictures/plot.eps")
```

```python
w = 0.8 / 1000  # m
h = 50 * 10**-9 # m
S = w*h         # m^2

l = 0.003       # m
dl = 0.05/1000  # m

dR = 0.0002
resistance_data = pd.read_csv("resistance_data.csv")
R = resistance_data["R"].to_numpy()
L = l * resistance_data["L"].to_numpy()

rhos = S * R / L
drhos = rhos * np.sqrt((dR/R)**2 + (dl/L)**2)

rhos =  10**9 *  rhos # Ohm * nm
drhos = 10**9 * drhos # Ohm * nm

rho = pd.DataFrame(columns=['pair', 'rho', 'drho'])
rho['rho'] = rhos
rho['drho'] = drhos
rho['pair'] = resistance_data['pair']
rho
```
