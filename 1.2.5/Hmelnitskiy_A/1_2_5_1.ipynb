```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from IPython.display import display, Math, Latex
import numpy as np

plt.rcParams["font.family"] = "monospace"

def mapping(x, k):
    return  x*k

plt.figure(figsize=(11,11))
plt.title(r'Зависимость угловой скорости рецессии $\Omega$ от момента силы $M$')
plt.ylabel(r'$\Omega$')
plt.xlabel("M")

#-------------------------------------------------------------------------------------
x = []
y = []

file = open('1.txt', 'r')
while (line := file.readline()):
  s = line.split()
  if (len(s) != 0):
    x.append(float(s[0].replace(',','.')))
    y.append(float(s[1].replace(',','.')))
k = 0
coeffs,_ = curve_fit(mapping, x, y)
k = coeffs
y_fit = []
for i in range(len(x)):
  y_fit.append(k * x[i])

plt.plot(x, y, 'r^', label=r'$Угловая скорость рецессии \Omega$ от момента силы $M$')
plt.plot(x, y_fit, color = 'b', label = 'График(с аппроксимацией)  ' r'$\Omega = M \cdot k$' ' , ' r'$k$' f' = {k[0]:.3f}')
plt.grid(visible = True, which='major', axis='both', alpha=1)
plt.grid(visible = True, which='minor', axis='both', alpha=1)
plt.legend()
plt.show()
plt.savefig('graphic.png')
file.close()
```
