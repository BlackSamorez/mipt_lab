```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
m = np.linspace(-4, 4, 9)
xm = np.array([0.04, 0.06, 0.087, 0.112, 0.14, 0.168, 0.19, 0.214, 0.24]) * 10
```

```python
plot_1 = plt.figure(figsize=(12*0.9,10*0.9))
plt.grid(visible=True, linewidth=0.6)

plt.title('Зависимость $x(m)$', fontsize=18)
#plt.xlim(xmin=0, xmax=6)
#plt.ylim(ymin=0, ymax=19)
plt.xlabel('$m$, порядок минимума', fontsize=20, rotation=0,
           #ha='right'
          )
plt.ylabel('$x$, мм', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=16, size=10)
plt.grid(visible=True, linewidth=0.6)


yerr_1 = 0.1
#xerr_1 = 0.01

#z, err = np.polyfit(D, Q[1], deg=1, cov=True)
#polynom_2 = np.poly1d(z)
#plt.plot(D, polynom_2(D),"b--", linewidth=1, label="МНК")

plt.errorbar(
    m,
    xm,
    fmt='ro',
    linewidth=1,
    markersize=4,
    elinewidth=1,
    label = 'Координата минимума от его порядка',
    #label='В "одну" сторону',
    #xerr=xerr_1,
    yerr=yerr_1,
    capsize=1
)

z = np.polyfit(m, xm, deg=1)
polynom_1 = np.poly1d(z)
plt.plot(m, polynom_1(m),"k--", linewidth=1, label="МНК")


plt.legend(fontsize=18, markerscale = 2)
plt.show()

plot_1.savefig(
    'x(m).pdf',
    format='pdf',
    bbox_inches='tight',
    #pad_inches=4
)
```
