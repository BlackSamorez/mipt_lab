```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_excel("graph_data.xlsx")
N = len(data['i'])
xerror = np.ones(N) * 0.05
yerror = data['n'] * 0.1
```

```python
fig, ax = plt.subplots()
ax.errorbar(data['i'], data['n'], yerr=yerror, xerr=xerror, capsize=2, color='black', fmt='.')
plt.plot(data['i'], data['n'], color='b')
ax.set_xlabel(r'$I$, A')
ax.set_ylabel(r'$N$, c$^{-1}$')
ax.grid(which='both', visible=True)


plt.savefig('curr_plot.png')
plt.show()
```
