```python
import matplotlib.pyplot as plt
#from sklearn.linear_model import LinearRegression
from scipy.optimize import curve_fit
import numpy as np
import matplotlib as mpl
from math import *

def mapping(x, k, b):
    return  k * x + b

plt.figure(figsize=(10,10))
plt.title(r"Зависимость $u(v)$ периода колебаний от положения груза")
plt.ylabel("u = T^2*xц, с^2*м")
plt.xlabel("v = y^2, м^2")

x = []
y = []

file = open('data2.txt', 'r')
while (line := file.readline()):
  s = line.split()
  y.append(float(s[0].replace(',','.')))
  x.append(float(s[1].replace(',','.')))

print(x)
print(y)

k = 0
b = 0
coeffs,_ = curve_fit(mapping, x, y)
k, b = coeffs
y_fit = []
for i in range(len(x)):
  y_fit.append(k * x[i] + b)
print(k, b)
plt.plot(x, y, 'r^', label='u(v)')
plt.plot(x, y_fit, label="Аппроксимирующая прямая y = kx + b, k = 0.9944, b = 0.48")

plt.grid(visible = True, which='major', axis='both', alpha=1)
plt.grid(visible = True, which='minor', axis='both', alpha=1)

plt.legend()
plt.show()
plt.savefig('graphic.png')
file.close()
```
