```python
import matplotlib.pyplot as plt
#from sklearn.linear_model import LinearRegression
from scipy.optimize import curve_fit
import numpy as np
import matplotlib as mpl
from math import *

def mapping(x, k, b, h):
    return  k * ((b + x**2)/(h + x))**0.5

plt.figure(figsize=(10,10))
plt.title(r" $T(y)$     ")
plt.ylabel("T, ")
plt.xlabel("y, ")

x = []
y = []

file = open('data.txt', 'r')
while (line := file.readline()):
  s = line.split()
  y.append(float(s[0].replace(',','.')))
  x.append(float(s[1].replace(',','.')))

print(x)
print(y)

k = 0
b = 0
coeffs,_ = curve_fit(mapping, x, y)
k, b, h = coeffs
y_fit = []
for i in range(len(x)):
  y_fit.append(k * ((b + x[i]**2)/(h + x[i]))**0.5)

plt.plot(x, y, 'r^', label='    ')
plt.plot(x, y_fit, label="y = k * ((b + x**2)/(h + x))**0.5 ")

plt.grid(visible = True, which='major', axis='both', alpha=1)
plt.grid(visible = True, which='minor', axis='both', alpha=1)

plt.legend()
plt.show()
plt.savefig('graphic.png')
file.close()
```
