```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
d = np.array([
10,
24.5,
49.7])

D = np.array([
4,
2.21,
1.03])
```

```python
plot_1 = plt.figure(figsize=(12*0.9,10*0.9))
plt.grid(visible=True, linewidth=0.6)

plt.title('Зависимость $d(1/D)$', fontsize=18)
#plt.xlim(xmin=0, xmax=6)
plt.ylim(ymin=0, ymax=55)
plt.xlabel('$1/D$, мм$^{-1}$', fontsize=20, rotation=0, ha='right')
plt.ylabel('$d$, мкм', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=16, size=10)
plt.grid(visible=True, linewidth=0.6)



#yerr_1 = 0.05
#xerr_1 = 0.01


#z, err = np.polyfit(D, Q[1], deg=1, cov=True)
#polynom_2 = np.poly1d(z)
#plt.plot(D, polynom_2(D),"b--", linewidth=1, label="МНК")

plt.errorbar(
    1/D,
    d,
    fmt='ro',
    linewidth=1,
    markersize=8,
    elinewidth=1,
    label='d(1/D)',
    #xerr=xerr_1,
    #yerr=yerr_1,
    capsize=1
)

z = np.polyfit(1/D[1:], d[1:], deg=1)
polynom_1 = np.poly1d(z)
plt.plot([0,1], polynom_1([0,1]),"k--", linewidth=1, label="МНК")

plt.annotate('Это выброс',
            xy=(1/D[:-1][0], d[:-1][0]),
            xytext=(1/D[:-1][0], d[:-1][0]+1),
            )


plt.legend(fontsize=18, markerscale = 1)
plt.show()

plot_1.savefig(
    'd.pdf',
    format='pdf',
    bbox_inches='tight',
    #pad_inches=4
)
```
