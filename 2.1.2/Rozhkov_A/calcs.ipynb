```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
from scipy.optimize import curve_fit
from lab_lib import *
```

```python
SIGMA_DH = 0.2
SIGMA_T  = 0.3
```

```python
# 1 пункт

h1_1 = [10.7, 10.5, 10.3, 10.1, 9.9, 9.7, 9.5,  9.3]
t_1  = [0,    5,    5.8,  6.8,  7.8, 8.9, 17.7, 35]

table_1 = [["$t, с$", "$h_1, см.вод.ст.$"]] + [[t_1[i], h1_1[i]] for i in range(len(h1_1))]

texTable(table_1)

make_raw_plot("img/plot_h1_t_1.png", "График зависимости $h_1(t)$", "$t, с$", "$h_1, см.вод.ст.$", t_1, h1_1, 0.3, 0.2)
```

```python
def make_gamma_ans_from_res(D_t_p, res):
    for i in range(len(res)):
        h1 = res[i][0]
        h2 = res[i][1]
        gamma = h1 / (h1 - h2)
        res[i].append([gamma, SIGMA_DH * gamma / (h1 - h2) * np.sqrt((h2 / h1)**2 + 1)])

    table = [["$h_1, см.вод.ст.$", "$h_2, см.вод.ст.$", "$\gamma$"]] + [[res[i][0], res[i][1], texAns(res[i][2])] for i in range(len(res))]
    texTable(table)

    mean = meanWithInacc([i[2] for i in res])
    print("mean gamma (D_t_p = {}):".format(D_t_p), texAns(mean))
    return [[D_t_p, SIGMA_T], mean]

gamma = [0] * 6
```

```python
res_t_1 = [[9.3, 2],
           [9.3, 2],
           [9.5, 2.2],
           [9.5, 2.2],
           [9.3, 2],
           [9.5, 2.2],
           [9.5, 2.4],
           [9.5, 2.2],
           [9.3, 2.2],
           [9.5, 2.4]]

gamma[0] = make_gamma_ans_from_res(0.5, res_t_1)
```

```python
res_t_2 = [[9.5, 2.2], [9.5, 2.4]]

gamma[1] = make_gamma_ans_from_res(2, res_t_2)
```

```python
res_t_3 = [[9.5, 2.2], [9.3, 2.0]]

gamma[2] = make_gamma_ans_from_res(4, res_t_3)
```

```python
res_t_4 = [[9.1, 1.8], [9.5, 2.0]]

gamma[3] = make_gamma_ans_from_res(6, res_t_4)
```

```python
res_t_5 = [[9.5, 1.8], [9.1, 1.8]]

gamma[4] = make_gamma_ans_from_res(8, res_t_5)
```

```python
res_t_6 = [[9.3, 1.8], [9.3, 1.6]]

gamma[5] = make_gamma_ans_from_res(10, res_t_6)
```

```python
x  = [i[0][0] for i in gamma]
dx = [i[0][1] for i in gamma]

y  = [i[1][0] for i in gamma]
dy = [i[1][1] for i in gamma]

mpl.rcParams['font.size'] = 16                   # Управление стилем, в данном случаем - размером шрифта
plt.figure(figsize = (10,8), facecolor = "white") # Создаем фигуру

# Подписываем оси и график
plt.title("График зависимости $\gamma(\Delta t_p)$")
plt.ylabel("$\gamma$")
plt.xlabel("$\Delta t_p, с$")

def func(x, k, b):
    return x * k + b

popt, pcov = curve_fit(func, x, y, p0 = (0.0, 0.0))
k, b = popt
dk, db = np.sqrt(np.diag(pcov))

print("k: ({} +- {})".format(k, dk))
print("b: ({} +- {})".format(b, db))

#plt.errorbar(x, y, "or", markersize = 9, label = 'Экспериментальные значения')
plt.errorbar(x, y, xerr = dx, yerr = dy, fmt='.r', label = 'Экспериментальные значения')
plt.errorbar(0, b, fmt = "xg", label = "Экстраполяция к $\Delta t_p = 0$", markersize = 15, elinewidth = 5)

x_lin = np.linspace(0, x[-1], 1000)
plt.plot(x_lin, func(x_lin, k, b), "b", label = "Аппроксимация")

plt.grid(visible = True, which = 'major', axis = 'both', alpha = 1, linewidth = 0.9)   # Активируем сетку
plt.grid(visible = True, which = 'minor', axis = 'both', alpha = 0.5, linestyle = ':')

plt.minorticks_on()
plt.tight_layout()
plt.legend(loc = "best", fontsize = 12) # Активируем легенду графика

plt.savefig("{}".format("img/plot_gamma_t.png"))
plt.show()

gamma_ans = [b, np.sqrt(np.mean(dy)**2 + (k * SIGMA_T)**2 + dk**2 * np.mean(x)**2)]
print("gamma:", texAns(gamma_ans))
```
