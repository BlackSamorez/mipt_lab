# Определение пластин $\lambda / 2$, $\lambda / 4$

Условно скажем что с кружочком [ (4) ] -- 1 пластина, без кружочка [ 4 ] -- 2 пластина

1 пластинка : От минимума до минимума -- четверть оборота
минимум в $86\pm2^\circ$, $358\pm2^\circ$, $266\pm2^\circ$, $178\pm2^\circ$

2 пластинка : От минимума до минимума -- четверть оборота
минимум в $322\pm2^\circ$, $232\pm2^\circ$, $144\pm2^\circ$, $54\pm2^\circ$

совпадение направлений борелизаторов и пластин -- когда минимум интенсивности

1 пластинка -- меняется цвет, интенсивность почти одинакова -> $\lambda/4$, круговая борелизация

2 пластинка -- интенсивность достигает ярко выраженного минимума -> $\lambda/2$, линейная борелизация

```python
import matplotlib.pyplot as plt
import numpy as np

import math
import colour
```

```python
def curve(param, pi):
    return np.cos(pi * math.pi * 560/param)**2, np.sin(pi * math.pi * 560/param)**2

z = 5/4

ls = np.arange(400, 700, 1)
wl = ls


perm = curve(ls, z)[0]
sd = colour.SpectralDistribution(perm, wl, name = 'Parallel')

xyz = colour.sd_to_XYZ(sd, illuminant=colour.SDS_ILLUMINANTS["D65"]) / 100
colour.plotting.plot_single_colour_swatch(colour.XYZ_to_sRGB(xyz))

perm = curve(ls, z)[1]
sd2 = colour.SpectralDistribution(perm, wl, name='Perpendicular')

xyz = colour.sd_to_XYZ(sd2, illuminant=colour.SDS_ILLUMINANTS["D65"]) / 100
colour.plotting.plot_single_colour_swatch(colour.XYZ_to_sRGB(xyz))
colour.plotting.plot_single_sd(sd2)
```

```python

plt.plot(ls, curve(ls, z/2), color='black')

ls = np.linspace(380, 450, 100)
plt.fill_between(ls, np.zeros(100), curve(ls, z/2), color='magenta')
ls = np.linspace(450, 500, 100)
plt.fill_between(ls, np.zeros(100), curve(ls, z/2), color='blue')
ls = np.linspace(500, 520, 100)
plt.fill_between(ls, np.zeros(100), curve(ls, z/2), color='cyan')
ls = np.linspace(520, 565, 100)
plt.fill_between(ls, np.zeros(100), curve(ls, z/2), color='lawngreen')
ls = np.linspace( 565,590, 100)
plt.fill_between(ls, np.zeros(100), curve(ls, z/2), color='yellow')
ls = np.linspace(590,635, 100)
plt.fill_between(ls, np.zeros(100), curve(ls, z/2), color='orange')
ls = np.linspace(635, 770, 100)
plt.fill_between(ls, np.zeros(100), curve(ls, z/2), color='red')
```
