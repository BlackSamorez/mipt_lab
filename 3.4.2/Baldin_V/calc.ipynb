```python
data = pd.read_excel('data.xlsx', skiprows=3)
```

```python
T_0 = 9.045
```

```python
tau = data['FC']
ydata = [1 / (t**2 - T_0**2) for t in tau]
yerr = [ydata[i] * np.sqrt((0.02 / (tau[i] + T_0))**2 + (0.02 / (tau[i] - T_0))**2) for i in range(len(tau))]
temp = data['T']
```

```python
plt.figure(figsize=FIGSIZE)

CUT_IDX = 4
[a, b] = plot_linear(temp[CUT_IDX:], ydata[CUT_IDX:], xerr=0.5, yerr=yerr[CUT_IDX:],
                     datalabel='', xmin=13, xmax=max(temp))
plt.errorbar(temp[:CUT_IDX], ydata[:CUT_IDX], xerr=0.5, yerr=yerr[:CUT_IDX], fmt='k.')
plt_singleton(title='', xlabel=r'$t_{\text{}},\ ^{\circ}C$', ylabel=r'$\frac{1}{\tau^2 - \tau_0^2}$')
plt.ylim(bottom=0)
plt.savefig('graph.pdf')
```
