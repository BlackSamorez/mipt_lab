```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Mean squared error
x = np.array([1, 2, 3])
y = np.array([1, 2, 1])
n = 8
k = (np.mean(x * y) - np.mean(x) * np.mean(y)) / (np.mean(x * x) - np.mean(x) ** 2)
b = np.mean(y) - k * np.mean(x)
k_o = np.mean(x*y) / np.mean(x ** 2)
error_k = 1 / np.sqrt(n) * np.sqrt(((np.mean(y * y)) - np.mean(y) ** 2) / (np.mean(x * x) - np.mean(x) ** 2) - k**2)
k, error_k, b
```

```python
torr_40 = pd.read_csv('40tor.csv')
torr_100 = pd.read_csv('100tor.csv')
torr_160 = pd.read_csv('160tor.csv')
torr_220 = pd.read_csv('220tor.csv')
torr_280 = pd.read_csv('280tor.csv')
```

```python
def convert(dataframe):
    dataframe = dataframe.to_numpy()
    dataframe = dataframe.reshape(-1)
    return dataframe[::2], dataframe[1::2]
```

```python
torr_40_time, torr_40_volt = convert(torr_40)
torr_100_time, torr_100_volt = convert(torr_100)
torr_160_time, torr_160_volt = convert(torr_160)
torr_220_time, torr_220_volt = convert(torr_220)
torr_280_time, torr_280_volt = convert(torr_280)
```

```python
plot_1 = plt.figure(figsize=(12*0.9,10*0.9))
plt.grid(visible=True, linewidth=0.6)

plt.title(' $U(t)$', fontsize=18)
plt.xlim(xmin=0, xmax=400)
#plt.ylim(ymin=0, ymax=140)
plt.ylabel('$U$, ..', fontsize=20, rotation=0, ha='right')
plt.xlabel('$t$, ', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=16, size=10)
plt.grid(visible=True, linewidth=0.6)



#yerr_1 = pressure_sub * 0.005
#xerr_1 = cons * 0.01

#z = np.polyfit(D, Q[0], deg=1)
#polynom_1 = np.poly1d(z)
#plt.plot(D, polynom_1(D),"r--", linewidth=1, label="")

#z, err = np.polyfit(D, Q[1], deg=1, cov=True)
#polynom_2 = np.poly1d(z)
#plt.plot(D, polynom_2(D),"b--", linewidth=1, label="")

plt.errorbar(
    torr_40_time,
    torr_40_volt,
    fmt='r.',
    linewidth=0,
    markersize=2,
    elinewidth=1,
    label='40 torr',
    #xerr=xerr_1,
    #yerr=yerr_1
)
plt.errorbar(
    torr_100_time,
    torr_100_volt,
    fmt='g.',
    linewidth=0,
    markersize=2,
    elinewidth=1,
    label='100 torr',
    #xerr=xerr_1,
    #yerr=yerr_1
)
plt.errorbar(
    torr_160_time,
    torr_160_volt,
    fmt='b.',
    linewidth=0,
    markersize=2,
    elinewidth=1,
    label='160 torr',
    #xerr=xerr_1,
    #yerr=yerr_1
)
plt.errorbar(
    torr_220_time,
    torr_220_volt,
    fmt='k.',
    linewidth=0,
    markersize=2,
    elinewidth=1,
    label='220 torr',
    #xerr=xerr_1,
    #yerr=yerr_1
)
plt.errorbar(
    torr_280_time,
    torr_280_volt,
    fmt='y.',
    linewidth=0,
    markersize=2,
    elinewidth=1,
    label='280 torr',
    #xerr=xerr_1,
    #yerr=yerr_1
)


plt.legend(fontsize=18, markerscale = 5)
plt.show()

plot_1.savefig(
    'u(t).pdf',
    format='pdf',
    bbox_inches='tight',
    #pad_inches=4
)
```
