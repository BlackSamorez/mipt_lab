# Создание поля для графика

Для начала импортируем библиотеку

```python
from lab_cheat import *
```

Структура работы данного блока устроена так, что все элементы будут наноситься на объект класса __Figure__, сам объект преставляет собой полотно графика, оси и их названия. Далее представлены параметры, которые нужно указывать при создании полотна.

### Основные параметры:
- __x_label__: Название оси X.
- __y_label__: Название оси Y.
- __zero_in_corner__: True - точка (0, 0) отобразится на графике, и в ней будут пересекаться оси X и Y, если False, то программа сама настроит масштаб (Использовать True, если нужно показать точку пересечения с осями). _(Изначально True)_
- __label_near_arrow__: Если True: Названия будут отображаться в углу рядом с концами стрелок осей, если False: Названия будут отображаться около середин осей. _(Изначально True)_

### Редкоиспользуемые параметры:
- __x_label_coords__: Координаты местоположения названия оси X, следует использовать только если название отображено некорректно.
- __y_label_coords__: Аналог __x\_label\_coords__.
- __bold_axes__: Если True - то оси будут жирными, иначе будут как обычные прямые на графике.
- __legend_props__: Способ ручного контроля легенды, с его помощью на самом низшем и некрасивом уровне дать вашим прямым названия.
- __my_func__: Данный параметр отвечает за отрисовку вашего особого объекта на графике, будь то табличка и ещё что. Используется если в библиотеке не хватает функционала для ваших нужд.

```python
fig = Figure(x_label='X, м', y_label='Y, с')
fig.show()
```

# Функция line

С помощью библиотеки matplotlib отрисовывается прямая вида kx + b с заданными коэффициентами k и b. 

__Важное замечание:__ прямая, построенная функцией line(), является вспомогательным объектом, то есть не настраивает под себя оси. Если вы хотите, чтобы прямая настроила под себя оси воспользуйтесь функцией [_func_graph_](#Функция-func_graph) с параметром add_before_fixing_axes = False.


### Обязательные параметры:
- __k__ - Коэффициент наклона прямой
- __b__ - Свободный член прямой


### Дополнительные параметры:
- __colour__: Задаёт цвет желаемой прямой, поддерживает формат HEX ("#f1f1f1"), так же есть встроенные упрощённые цвета:
    - <span style="color:blue">_"b"_ - __Синий__</span>.
    - <span style="color:green">_"g"_ - __Зелёный__</span>
    - <span style="color:red">_"r"_ - __Красный__</span>
    - <span style="color:cyan">_"c"_ - __Бирюзовый__</span>
    - <span style="color:magenta">_"m"_ - __Маджента__</span>
    - <span style="color:yellow">_"y"_ - __Жёлтый__</span>
    - <span style="color:black">_"k"_ - __Чёрный__</span>
    - <span style="color:gray">_"w"_ - __Белый (написано серым (не белым))__</span>


- __line_style__: Позволяет настроить тип постоенной прямой, сомые распространённые стили:
    - _"solid"_ - Сплошная линия
    - _"dotted"_ - Линия "точками"
    - _"dashed"_ - Пунктирная линия (палочками)
    - _"dashdot"_ - Линия штрих-пунктирная линия
    - [_Больше стилей_](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)
    
    
- __label__: Позволяет назвать вашу прямую для аккуратного отображения в легенде

```python
k = 1.5
b = -0.3
colour = 'm'
line_style = 'dashed'
label = 'Example line'

fig.line(k, b, colour, line_style, label)
fig.show()
```

## Функция mnk

Функция реализует Метод наименьших квадратов. Считает два вида ошибок: вызываемый погрешностями и вызываемый статистикой.
Аппроксимация будет заисеть от того, насколько "хорошо" точки ложаться на график. Если точки "хорошо" ложатся на прямую, то преобладать будет ошибка из-за погрешностей. Если точки измерены крайне точно, но на прямую они ложатся не достаточно корректно, то преобладает статистическая ошибка. Результирующей ошибкой выдаётся корень из суммы квадратов двух видов этих ошибок. 


### Обязательные параметры (mnk):

- __x__ - Итерируемый объект с абсциссами точек, рекомендован `GroupVar`


- __y__ - Итерируемый объект с ординатами точек, рекомендован `GroupVar`


### Дополнительные параметры (mnk):

- __figure__ - Объект класса Figure, он же наше полотно, передаётся если мы хотим, чтобы эта прямая была сразу построена на графике

- __colour__: Задаёт цвет желаемой прямой, ПОддерживает формат HEX ("#f1f1f1"), так же есть встроенные упрощённые цвета:
    - <span style="color:blue">_"b"_ - __Синий__</span>.
    - <span style="color:green">_"g"_ - __Зелёный__</span>
    - <span style="color:red">_"r"_ - __Красный__</span>
    - <span style="color:cyan">_"c"_ - __Бирюзовый__</span>
    - <span style="color:magenta">_"m"_ - __Маджента__</span>
    - <span style="color:yellow">_"y"_ - __Жёлтый__</span>
    - <span style="color:black">_"k"_ - __Чёрный__</span>
    - <span style="color:gray">_"w"_ - __Белый (написано серым (не белым))__</span>


- __line_style__: Позволяет настроить тип постоенной прямой, сомые распространённые стили:
    - _"solid"_ - Сплошная линия
    - _"dotted"_ - Линия "точками"

    - [_Больше стилей_](https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)
    
    
- __label__: Позволяет назвать вашу прямую для аккуратного отображения в легенде

## Функция mnk_through0

Данная функция ведёт себя аналогично функции [__mnk__](#Функция-mnk), но проводит прямую через начало координат.
