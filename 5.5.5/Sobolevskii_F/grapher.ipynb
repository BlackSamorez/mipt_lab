```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

background = pd.read_csv("background_raw.csv")

fig, ax = plt.subplots()

plt.plot(background['Channel'], background['Counts'])
ax.grid()
ax.set_xlabel(r'Номер канала $N$')
ax.set_ylabel(r'Количество зафиксированных частиц')

plt.savefig('background.png')
plt.show()
```

```python
Co60 = pd.read_csv('Co60_raw.csv')
Na22 = pd.read_csv('Na22_raw.csv')
Cs137 = pd.read_csv('Cs137_raw.csv')
Eu152 = pd.read_csv('Eu152_raw.csv')
Am241 = pd.read_csv('Am241_raw.csv')

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].plot(Co60['Channel'], Co60['Counts'])
ax[1].plot(Na22['Channel'], Na22['Counts'])

for i in range(2):
    ax[i].grid()
    ax[i].set_ylim((0, 150))
    ax[i].set_xlabel(r'Номер канала $N$')

ax[0].set_title(r'$^{60}$Co')
ax[1].set_title(r'$^{22}$Na')
ax[0].set_ylabel('Количество частиц')

plt.savefig('Na22Co60.png')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(Cs137['Channel'], Cs137['Counts'])

ax.grid()
ax.set_xlabel(r'Номер канала $N$')
ax.set_ylabel(r'Количество частиц')

plt.savefig('Cs137.png')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].plot(Eu152['Channel'], Eu152['Counts'])
ax[1].plot(Am241['Channel'], Am241['Counts'])

for i in range(2):
    ax[i].grid()
    ax[i].set_ylim((0, 4500))
    ax[i].set_xlim((0, 600))
    ax[i].set_xlabel(r'Номер канала $N$')

ax[0].set_title(r'$^{152}$Eu')
ax[1].set_title(r'$^{241}$Am')
ax[0].set_ylabel('Количество частиц')

plt.savefig('Eu152Am241.png')

plt.show()
```

```python
Ns = [1592, 1804, 729, 1720, 926]
Es = [1.173, 1.332, 0.511, 1.274, 0.662]
xerror = np.ones(5)
yerror = np.ones(5) * 0.002

grid = np.arange(50, 2000, 1)

fig, ax = plt.subplots(figsize=(5, 4))

ax.errorbar(Ns, Es, yerr=yerror, xerr=xerror, capsize=2, color='black', fmt='.')
plt.plot(grid, 0.000767 * grid - 0.04786)
ax.grid()
ax.set_ylabel(r'$E$, МэВ')
ax.set_xlabel(r'Номер канала $N$')

plt.savefig('calibr.png')
plt.show()
```

```python
data = pd.read_csv('graph2.csv')

xerror = np.ones(8) * 0.001
yerror = np.ones(8) * 2

grid = np.arange(0, 0.11, 0.001)

fig, ax = plt.subplots(figsize=(5, 4))

ax.errorbar(data['e'], data['r'], yerr=yerror, xerr=xerror, capsize=2, color='black', fmt='.')
plt.plot(grid, 238 * grid)
ax.grid()
ax.set_ylabel(r'$R^2$')
ax.set_xlabel(r'$1/E$, МэВ$^{-1}$')

plt.savefig('graph2.png')
plt.show()
```
