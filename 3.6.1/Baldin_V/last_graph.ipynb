```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv('last.csv')
nu = df['n'] * 300
K = df['K']
time = 1 / nu

plt.figure(figsize=(12,6))
plt.errorbar(time, K, fmt='b.')
x = np.linspace(min(time), max(time), 1000)
[a, b] = np.polyfit(time, K, deg=1)
plt.plot(x, a*x + b)

plt.xlabel('$t$, мс')
plt.ylabel('$K$')
plt.title('График зависимости $K(t)$')
plt.grid(which='minor', linestyle='--')
plt.grid(which='major', linestyle='-')
plt.minorticks_on()

plt.savefig('photo/K(nu).png')
```
